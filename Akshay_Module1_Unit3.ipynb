```python
India_runs.size
```

```python
matches = np.arange(10)
matches
```

```python
theta1 = np.arange(360)
theta1_r = np.deg2rad(theta1)
sintheta = np.sin(theta1_r)
plt.plot(theta1,sintheta)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sine wave")
plt.show()
```

```python
theta1 = np.arange(360)
theta1_rad = np.deg2rad(theta1)
costheta = np.cos(theta1_rad)
plt.plot(theta1,costheta)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Cosine Wave")
```

```python
theta1 = np.arange(360)
theta1_rad = np.deg2rad(theta1)
sintheta = np.sin(theta1_r)
costheta = np.cos(theta1_rad)

plt.figure(figsize=(20,20))

plt.plot(theta1,sintheta)
plt.plot(theta1,costheta)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.annotate("Maxima",xy=(90,1))
plt.legend(['Sine Wave','Cos Wave'])
plt.title("Sine and Cosine Wave")
plt.savefig("Figure.pdf")
plt.show()
```
